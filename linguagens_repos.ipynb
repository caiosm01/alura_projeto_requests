### Autentica√ß√£o

```python
import os
from dotenv import load_dotenv


#load_dotenv(override=True)
access_token = os.getenv("access_token")

# Especifincando a vers√£o da API
headers = {'X-GitHub-Api-Version': '2022-11-28',
           'Authorization': 'Bearer '+ access_token}
```

### Paginando os reposit√≥rios

```python
api_base_url = 'https://api.github.com'
owner = 'caiosm01' # username de quem vamos extrair os dados
#owner = 'amzn'
url = f'{api_base_url}/users/{owner}/repos'
```

```python
repos_list = []

for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
len(repos_list[0])
```

## Transformando os dados

### Nomes dos reposit√≥rios

```python
repos_list[0][1]['name']
```

```python
repos_name = []

for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
repos_name
```

### Linguagens dos reposit√≥rios

```python
repos_list[0][0]['language']
```

```python
repos_language = []

for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
```

```python
repos_language
```

## Criando um DataFrame

```python
import pandas as pd
```

```python
dados = pd.DataFrame()
dados['repository_name'] = repos_name
dados['language'] = repos_language
```

```python
dados
```

```python
print(dados.head())
```

```python
dados.to_csv('dados.csv', index=False)
```

## Armazenando os dados

## Formato do arquivo

```python
import base64
```

```python
with open('dados.csv', 'rb') as file:
    file_content = file.read()

enconded_content = base64.b64encode(file_content)
```

## Upload de arquivo com PUT

```python
api_base_url = 'https://api.github.com'
username = 'caiosm01'
repo = 'linguagens-utilizadas'
path = 'dados.csv'

url = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'

url
```

```python
data = {
    'message': 'Adicionando um novo arquivo',
    'content': enconded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
response.status_code
```
